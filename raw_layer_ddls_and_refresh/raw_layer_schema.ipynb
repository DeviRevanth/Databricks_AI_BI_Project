```python
%sql

CREATE TABLE IF NOT EXISTS ecommerce.raw.orders (
  order_id STRING,
  order_ts STRING,
  customer_id STRING,
  product_id STRING,
  quantity INT,
  unit_price DOUBLE,
  line_amount DOUBLE,
  payment_method STRING,
  order_status STRING
)
USING CSV
OPTIONS (header = "true")
LOCATION 's3://databricks-ai-bi-project/raw/orders/';

CREATE TABLE ecommerce.raw.brands (
  brand_id STRING,
  brand_name STRING
)
USING CSV
OPTIONS (
  header = "true"
)
LOCATION 's3://databricks-ai-bi-project/raw/brands/';

CREATE TABLE ecommerce.raw.customers (
  customer_id STRING,
  customer_name STRING,
  email STRING,
  country STRING,
  phone_number STRING,
  gender STRING,
  created_at STRING
)
USING CSV
OPTIONS (
  header = "true"
)
LOCATION 's3://databricks-ai-bi-project/raw/customers/';


CREATE TABLE ecommerce.raw.categories (
  category_id STRING,
  category_name STRING
)
USING CSV
OPTIONS (
  header = "true"
)
LOCATION 's3://databricks-ai-bi-project/raw/categories/';

CREATE TABLE ecommerce.raw.products (
  product_id STRING,
  product_name STRING,
  brand_id STRING,
  category_id STRING,
  price DOUBLE,
  created_at STRING
)
USING CSV
OPTIONS (
  header = "true"
)
LOCATION 's3://databricks-ai-bi-project/raw/products/';
```
